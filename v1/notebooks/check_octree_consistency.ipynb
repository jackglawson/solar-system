```python
# check that python and cpp implementations are returning the same acceleration
```

```python
from solar_system import *
from params import *
from visualise import *
from octree_module import find_acc_cpp
from initial_conditions import generate_spherical_uniform_rs
```

```python
rs = generate_spherical_uniform_rs()
```

```python
len(rs)
```

```python
ms = np.array([1] * len(rs))
```

```python
#python brute force
s = time.time()
py_bf_acc = find_acc_v(rs, ms, len(rs))
f = time.time()
print(f-s)
```

```python
# python octree
centre = np.array([0,0,0])
width = np.max(np.abs(rs)) * 2 + 1
s = time.time()
py_octree = Octree(centre, width, rs, ms)
py_acc = py_octree.find_acc_v(rs, ms)
f = time.time()
print(f-s)
```

```python
#cpp octree
rs_tuple = tuple(map(tuple, rs))
ms_tuple = tuple(ms)
s = time.time()
cpp_acc = find_acc_cpp(rs_tuple, ms_tuple, (p.unit_mass_particle_radius, p.qt, p.G))
f = time.time()
print(f-s)
```

```python
cpp_acc = np.array(cpp_acc)
```

```python
fractional_diff = (cpp_acc - py_acc) / py_acc
```

```python
diff = cpp_acc - py_acc
```

```python
diff
```

```python
for i in range(len(rs)):
    print(fractional_diff[i])
    print(py_acc[i])
    print(cpp_acc[i])
    print("\n")
```
